## Advance Operation with Business Examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structures :[restaurant_id,2021,2022,2023,2024]
sales_data=np.array([
    [1,150000,180000,220000,250000], # Paradise Biryani
    [2,120000,140000,160000,190000], # Beijing Bites
    [3,200000,230000,260000,300000], # Pizza Hub
    [4,180000,210000,240000,270000], # Burger Point
    [5,160000,185000,205000,230000], # Chai Point
])

print("==== Zomato Sales analysis ==== ")
print(" \n Sales Data Shape :",sales_data.shape)
print("1st Three Restaurant :",sales_data[0:3])
```

### Total Sales Per Year

```python
print("Total Sales Per Year :",np.sum(sales_data[:,1:],axis=0))
print("Total Sales Per Restraunt :",np.sum(sales_data[:,1:],axis=1))
print("Total sales of all Years :",np.sum(sales_data[:,1:]))

# Minimum Sales Per Restaurant 
min_sales=np.min(sales_data[:,1:],axis=1)
print("Minimum Sales Per Restaurant :",min_sales)

# Maximum Sales Per Year
print("Maximum Sales per Year :",np.max(sales_data[:,1:],axis=0))

# Average Sales Per Restaurant 
avg_sales=np.mean(sales_data[:,1:],axis=1)
print("Average Sales Per Restaurant :",avg_sales)

# Cumalative Sales:
cumsum=np.cumsum(sales_data[:,1:],axis=1)
print("Cumalative Sales : \n",cumsum)

plt.figure(figsize=(6,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average Cumulative Sales Per Year Accross All Restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

### Vector Operation

```python
vector1 =np.array([1,2,3,4,5])
vector2=np.array([6,7,8,9,10])

print("Vector Addition :",vector1+vector2)
print("\nMuliplication Vector of Vector :",vector1*vector2)
print("\nDot Product of Vector :",np.dot(vector1,vector2))
angle=np.arccos((np.dot(vector1,vector2))/(np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print("\nAngle Between Vectors :",angle)
```
